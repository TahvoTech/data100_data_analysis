Question

Toteuta Python-ohjelma, joka saa syötteenään tuntikohtaista liikennedataa sisältävän CSV-tiedoston ja laskee sekä tulostaa kunkin tunnin arvojen keskiarvot. Lue tiedosto käyttäen Pythonin csv-modulia.

Syötetiedoston nimi annetaan komentoriviparametrina (katso koodipohjaa). Syötetiedoston oletetaan olevan samaa muotoa kuin viikon 1 alaisuudessa ladattavissa oleva tiedosto "liikenne1.csv". Tiedoston kukin tietorivi kuvaa eräällä tieosuudella mitatut tuntikohtaiset liikennemäärät (mittauspisteen ohittaneiden ajoneuvojen lukumäärät) yhden päivän kaikille tunneille. Vuorokauden 24 eri tuntia vastaavat sarakeotsikot ovat muotoa "AA_BB", missä AA esittää alkutunnin ja BB lopputunnin kahdella numerolla. Esimerkiksi ensimmäinen tunti on "00_01", toinen on "01_02" ja viimeinen on "23_24".

Ohjelmasi tulee tulostaa 24 riviä, joista kukin on muotoa "AA-BB: ka", missä ka on tuntia "AA_BB" koskeneiden ajoneuvolukumäärien keskiarvo kahden desimaalin tarkkuudella. Rivit tulostetaan tuntien luonnollisessa järjestyksessä tunnista 00-01 tuntiin 23-24. Huomaa, että rivien alut on muotoiltu hieman eri tavalla kuin sarakkeet (alku- ja lopputunnin välissä on alaviivan sijaan väliviiva).

HUOM! Tehtävän syötetiedostossa voi olla tyhjiä arvoja. Tarkistin olettaa, että tällaisia puuttuvia arvoja kohdellaan nollina: käytä siis keskiarvon laskennassa jakajana kaikkien tietorivien lukumäärää riippumatta siitä, puuttuiko arvoja. Tällaista puuttuvien arvojen tulkintaa nolliksi kutsutaan nimellä "zero imputation", ja se ei sinänsä välttämättä ole oikeasti järkevää.

```python
from sys import argv
import csv

if __name__ == "__main__":
    filename = 'liikenne1.csv'

    with open(filename, mode='r', newline='', encoding='utf-8-sig') as file:
        start_column = '00_01'
        num_columns = 24
        reader = csv.DictReader(file, delimiter=';')
        fieldnames = reader.fieldnames
        start_index = fieldnames.index(start_column)
        selected_columns = fieldnames[start_index:start_index + num_columns]
        data = {col: [] for col in selected_columns}

        for row in reader:
            for col in selected_columns:
                try:
                    data[col].append(float(row[col]))
                except ValueError as e:
                    print(f"ValueError for column {col} with value {row[col]}: {e}")
                    data[col].append(0)

        for col, values in data.items():
            mean_value = sum(values) / len(values) if values else 0
            formatted_col = col.replace('_', '-')
            print(f"{formatted_col}: {mean_value:.2f}")
```

```python
from sys import argv
import csv

if __name__ == "__main__":
    filename = 'liikenne1.csv'

    n = 0
    tunnit = [f"{h:02}_{h+1:02}" for h in range(24)]
    # tunnit = [] # this is the same as the line above
    # for h in range(24):
    #     tunnit.append(f"{h:02}_{h+1:02}")
    counts = {}
    with open(filename) as f:
        for row in csv.DictReader(f, delimiter=';'):
            n += 1
            for tunti in tunnit:
                if tunti not in counts:
                    counts[tunti] = 0
                counts[tunti] += int(row[tunti] if row[tunti] else 0)
    for tunti in tunnit:
        print(f"{tunti.replace("_", "-")}: {counts[tunti]/n:.2f}")

            

        
```
