Kirjoita Python-ohjelma, joka avaa kurssilla käsiteltyä muotoa olevan tiedoston "Fifa Worldcup 2022.csv", laskee kunkin maan maalikeskiarvon (keskimäärin kuinka monta maalia per ottelu) ja tulostaa maat sekä niiden maalikeskiarvot järjestettynä ensisijaisesti maalikeskiarvojen mukaiseen laskevaan järjestykseen ja toissijaisesti maiden kasvavaan aakkosjärjestykseen. Kukin maa ja maalikeskiarvo tulostetaan omalle rivilleen muodossa "maa keskiarvo" ja keskiarvo tulostetaan 2 desimaalin tarkkuudella.
```python
import csv
from collections import defaultdict

# Read the data from the CSV file
with open('Fifa Worldcup 2022.csv', newline='') as f:
    reader = csv.reader(f)
    next(reader)  # Skip the header row
    rows = list(reader)

# Initialize dictionaries to store goals and matches count
goals = defaultdict(int)
matches = defaultdict(int)

# Process each row to calculate total goals and matches for each country
for row in rows:
    country1, country2 = row[2], row[3]
    goals1, goals2 = int(row[5]), int(row[6])
    
    goals[country1] += goals1
    goals[country2] += goals2
    matches[country1] += 1
    matches[country2] += 1

# Calculate average goals per match for each country
average_goals = {country: goals[country] / matches[country] for country in goals}

# Sort the countries by average goals (descending) and then by country name (ascending)
sorted_countries = sorted(average_goals.items(), key=lambda x: (-x[1], x[0]))

# Print the results
for country, avg_goals in sorted_countries:
    print(f"{country} {avg_goals:.2f}")
```

```python
import csv
from collections import defaultdict

with open('Fifa Worldcup 2022.csv', newline='') as f:
    reader = csv.reader(f)
    next(reader)
    rows = list(reader)

# Initialize dictionaries to store goals and matches count
goals = defaultdict(int)
matches = defaultdict(int)

# Process each row to calculate total goals and matches for each country
for row in rows:
    country = row[2]
    try:
        goals_scored = int(row[5])
    except ValueError:
        print(f"Skipping row with invalid goal value: {row}")
        continue  # Skip rows with invalid goal values
    
    print(f"Processing row: {row}")
    print(f"Country: {country}, Goals Scored: {goals_scored}")
    
    goals[country] += goals_scored
    matches[country] += 1

# Print the goals and matches dictionaries
print("Goals dictionary:", goals)
print("Matches dictionary:", matches)

# Calculate average goals per match for each country
average_goals = {country: goals[country] / matches[country] for country in goals}

# Print the average goals dictionary
print("Average goals dictionary:", average_goals)

# Convert the dictionary to a list of tuples
average_goals_list = list(average_goals.items())

# Sort the countries by average goals (descending) and then by country name (ascending)
sorted_average_goals_list = sorted(average_goals_list, key=lambda x: (-x[1], x[0]))

# Print the sorted list
print("Sorted average goals list:", sorted_average_goals_list)

# Print the results
for country, avg_goals in sorted_average_goals_list:
    print(f"{country} {avg_goals:.2f}")
```
