```python
paikat = []
for el in xml.iter("toimipaikka"):
    nimiEl = el.find("nimi", "").strip()
    nroEl = el.find("postinumero", "").strip()
    paikat.append((nimi, nro))
paikat
```

```python
paikat = [(el.findtext("nimi","").strip(), 
           el.findtext("postinumero", "").strip())
          for el in xml.iter("toimipaikka")]
paikat
```

```python
paikat = list(filter(lambda p: p[0] and p[1], paikat))
```

```python
paikat.sort()
paikat
```

```python
paikat.sort(key=lambda p: p[1], reverse=True)
paikat.sort(key=lambda p: p[0])
paikat
```

```python
juuri = ET.Element("postitoimipaikat")
print(juuri)
for p in paikat:
    tpEl = ET.SubElement(juuri, "toimipaikka")
    nimiEl = ET.SubElement(tpEl, "nimi")
    nimiEl.text = p[0]
    nroEl = ET.SubElement(tpEl, "postinumero")
    nroEl.text = p[1]
ET.indent(juuri)
print(ET.tostring(juuri, encoding="unicode"))
ET.ElementTree(juuri).write("paikat.xml", encoding="utf8")
```
